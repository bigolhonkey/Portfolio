# Project 3: Poisonous mushrooms

In this project, you'll investigate properties of mushrooms. This classic dataset contains over 8000 observations, where each mushroom is described by a variety of features like color, odor, etc., and the target variable is an indicator for whether the mushroom is poisonous. Since all the observations are categorical, I've binarized the feature space. Look at the feature_names below to see all 126 binary names.

You'll start by running PCA to reduce the dimensionality from 126 down to 2 so that you can easily visualize the data. In general, PCA is very useful for visualization (though sklearn.manifold.tsne is known to produce better visualizations). Recall that PCA is a linear transformation. The 1st projected dimension is the linear combination of all 126 original features that captures as much of the variance in the data as possible. The 2nd projected dimension is the linear combination of all 126 original features that captures as much of the remaining variance as possible. The idea of dense low dimensional representations is crucial to machine learning!

Once you've projected the data to 2 dimensions, you'll experiment with clustering using KMeans and density estimation with Gaussian Mixture Models. Finally, you'll train a classifier by fitting a GMM for the positive class and a GMM for the negative class, and perform inference by comparing the probabilities output by each model.

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from matplotlib.colors import LogNorm
```

Load feature names.

```python
feature_names = []
with open('mushroom.map') as fmap:
    for line in fmap:
        [index, name, junk] = line.split()
        feature_names.append(name)

print('Loaded feature names:', len(feature_names))
```

Load data. The data is sparse in the input file, but there aren't too many features, so we'll use a dense representation, which is supported by all sklearn objects.

```python
X, Y = [], []

with open('mushroom.data') as fdata:
    for line in fdata:
        items = line.split()
        Y.append(int(items.pop(0)))
        x = np.zeros(len(feature_names))
        for item in items:
            feature = int(item.split(':')[0])
            x[feature] = 1
        X.append(x)

# Convert these lists to numpy arrays.
X = np.array(X)
Y = np.array(Y)

# Split into train and test data.
train_data, train_labels = X[:7000], Y[:7000]
test_data, test_labels = X[7000:], Y[7000:]

# Check that the shapes look right.
print(train_data.shape, test_data.shape)
```

(1) Run Principal Components Analysis on the data. Show what fraction of the total variance in the training data is explained by the first k principal components, for k in [1, 50].

```python
def P1():
### STUDENT START ###
    ## Define pca decomposition object with n=50 components to keep
    pca = PCA(n_components=50)
    pca.fit(train_data) # No need to transform since we are only reading back the explained variances for plotting
    plt.plot(np.arange(1,51),np.cumsum(pca.explained_variance_ratio_))
    plt.ylabel('Explained variance ratio') 
    plt.xlabel('First k principal components')
    plt.title('Fraction of total variance explained by the first k principal components')  
    plt.show() 
### STUDENT END ###

P1()
```

(2) PCA can be very useful for visualizing data. Project the training data down to 2 dimensions and plot it. Show the positive (poisonous) cases in blue and the negative (non-poisonous) in red. Here's a reference for plotting: http://matplotlib.org/users/pyplot_tutorial.html

```python
def P2():
### STUDENT START ###
    # Define pca object with 2 dimensions, fit and transform the data
    pca_2comp = PCA(n_components=2).fit_transform(train_data)
    # Generate a scatter plot for representation in 2 dimensions 
    poisonous = plt.scatter(pca_2comp[:,0][train_labels==1],pca_2comp[:,1][train_labels==1],facecolor='blue',marker=".")
    non_poisonous = plt.scatter(pca_2comp[:,0][train_labels==0],pca_2comp[:,1][train_labels==0],facecolor='red',marker=".")
    # Plot legends for scatterplot
    plt.legend((poisonous,non_poisonous),
           ('positive(poisonous)', 'negative(non-poisonous)'),
           scatterpoints=1,
           loc='upper left')
    plt.xlabel("1st principal comp score")
    plt.ylabel('2nd principal comp score')
    plt.title("Training data projection to 2 dimensional space")
    plt.show()
### STUDENT END ###

P2()
```

(3) Run KMeans with [1,16] clusters over the 2d projected data. Mark each centroid cluster and plot a circle that goes through the most distant point assigned to each cluster.

```python
def P3():
### STUDENT START ###
    pca_2comp = PCA(n_components=2).fit_transform(train_data)
    num_clusters = np.arange(1,17)
    fig,ax = plt.subplots(4,4, figsize=(20,24))
    # generate plot indeces for subplot grid
    plot_index = []
    for i in range(4):
        for j in range(4):
            plot_index.append((i,j))
    # Loop over all cluster values 
    for cluster in num_clusters:
        km = KMeans(n_clusters=cluster) # define the kMeans classifier 
        km_fit = km.fit(pca_2comp)  # compute k-means clusters 
        cluster_centers = km_fit.cluster_centers_ # point to the cluster centers
        clust_dist_spaces = km_fit.transform(pca_2comp) # transform the data to cluster-distance space
        center1,center2 = zip(*cluster_centers) # collect cluster center coordinates
        colors = plt.cm.Accent(np.linspace(0,1,cluster))
        for k,center in enumerate(cluster_centers):
            clust_indeces = np.where(km.labels_==k)
            # Get PCA co-ordinates of all the clusters
            coord1,coord2 = zip(*pca_2comp[clust_indeces])
            # Compute the radius of the kth cluster from the max_dists array
            max_dist = np.max(clust_dist_spaces[clust_indeces][:,k]) # get the radius as farthest point from center     
            #Plot scatter plot 
            ax[plot_index[cluster-1]].scatter(coord1,coord2,facecolor=colors[k,:],linewidth=0.1)
            # Generate circle with radius between cluster center and farthest point in cluster 
            circle = plt.Circle(center, max_dist, color='black',fill=False,lw=0.2)
            ax[plot_index[cluster-1]].add_patch(circle)
            ax[plot_index[cluster-1]].set_aspect('equal', 'datalim')
        #Plot the cluster center
        ax[plot_index[cluster-1]].plot(center1,center2,'x',color='black')
        ax[plot_index[cluster-1]].axis([-3,3,-3,3])
        ax[plot_index[cluster-1]].set_title("k_clusters="+str(cluster))
        for a in ax.flat:
            a.set(xlabel='First PCA Score', ylabel='Second PCA score')
    plt.show() 
### STUDENT END ###

P3()
```

\(4) Fit a Gaussian Mixture Model for the positive examples in your 2d projected data. Plot the estimated density contours as shown here: http://scikit-learn.org/stable/auto_examples/mixture/plot_gmm_pdf.html#example-mixture-plot-gmm-pdf-py. Vary the number of mixture components from 1-4 and the covariance matrix type ('spherical', 'diag', 'tied', 'full').

```python
def P4():
### STUDENT START ###
    mixture_components = range(1,5)
    covariance_matrix = ['spherical', 'tied', 'diag', 'full']
    # Plotting options
    fig,ax = plt.subplots(4,4, figsize=(12,12))
    # generate plot indeces for subplot grid
    plot_index = []
    for i in range(4):
        for j in range(4):
            plot_index.append((i,j))
    # Project data in 2D
    pca_2comp = PCA(n_components=2).fit_transform(train_data)
    positive = pca_2comp[train_labels==1] ## Select the rows with positive examples
    
    for idx1,mix_component in enumerate(mixture_components):
        for idx2,cov in enumerate(covariance_matrix):
            # Fit the gaussian mixture model
            clf = GaussianMixture(n_components=mix_component, covariance_type=cov)
            clf.fit(positive)
            # display predicted scores by the model as a contour plot
            x = np.linspace(-3, 4)
            y = np.linspace(-3, 4)
            X, Y = np.meshgrid(x, y)
            XX = np.array([X.ravel(), Y.ravel()]).T
            Z = -clf.score_samples(XX)
            Z_shape = int(np.sqrt(len(Z))),int(np.sqrt(len(Z)))
            Z = Z.reshape(Z_shape)
            ax[idx1,idx2].axis('equal')
            ax[idx1,idx2].scatter(positive[:,0],positive[:,1],facecolor="pink",marker="o",alpha=0.5,s=8,lw=0.3)
            ax[idx1,idx2].axis("equal")
            ax[idx1,idx2].contour(X, Y, Z, norm=LogNorm(vmin=1.0, vmax=1000.0), levels=np.logspace(0, 2, 10))
            ax[idx1,idx2].set_xticks([])
            ax[idx1,idx2].set_yticks([])
            if idx1 is 0:
                ax[idx1,idx2].set_xlabel(covariance_matrix[idx2].upper())
                ax[idx1,idx2].xaxis.set_label_position('top') 
            if idx2 is 0:
                ax[idx1,idx2].set_ylabel('No. of components: {mc}'.format(mc=idx1+1))
    # remove spacing between plots in x and y
    fig.tight_layout()
    plt.subplots_adjust(wspace=0,hspace=0)  
    plt.show()
         
    
### STUDENT END ###

P4()
```

(5) Fit two 4-component full covariance GMMs, one for the positive examples and one for the negative examples in your 2d projected data. Predict the test examples by choosing the label for which the model gives a larger probability (use GaussianMixture.score_samples). What is the accuracy?

```python
def P5():
### STUDENT START ###
    # PCA for 2 dimensions
    pca_2comp = PCA(n_components=2)
    # Also transform train and tests sets
    pca_2comp_train = pca_2comp.fit_transform(train_data)
    pca_2comp_test = pca_2comp.transform(test_data)
    # get positive and negative examples
    positive = pca_2comp_train[train_labels==1]
    negative = pca_2comp_train[train_labels==0]
    # fit GMM for both positive and negative examples
    GMM = GaussianMixture(n_components=4, covariance_type='full')

    GMM_pos = GMM.fit(positive)
    positive_score = GMM_pos.score_samples(pca_2comp_test)
    GMM_neg = GMM.fit(negative) 
    negative_score = GMM_neg.score_samples(pca_2comp_test)
   
    '''Prediction algorithm:
        Predicted label positive (1) if positive is greater than negative
        Predict label negative (0) otherwise
    '''
    predictions  = positive_score > negative_score
    
    # Calculate the prediction accuracy
    accuracy = (np.sum(test_labels==predictions))/len(test_labels)
    print('Accuracy of combined GMM predicitons:{ac}'.format(ac=round(accuracy,4)))
### STUDENT END ###

P5()
```

(6) Ideally, we'd like a model that gives the best accuracy with the fewest parameters. Run a series of experiments to find the model that gives the best accuracy with no more than 50 parameters. For example, with 3 PCA components and 2-component diagonal covariance GMMs, you'd have:

( (3 mean vector + 3 covariance matrix) x 2 components ) x 2 classes = 24 parameters

You should vary the number of PCA components, the number of GMM components, and the covariance type.

```python
def P6():
### STUDENT START ###
    
    def fitGMM(cov,pca,gmm,rand):
        ''' 
        1. Reduce dimensionality with PCA 
        2. Fit a GMM with covariance and mixture components
        3. Calcualte negative log probability score with positive and negative samples
        4. Calculate accurancy for test label predictions
        '''
        pcan = PCA(n_components=pca)
        pcan_train = pcan.fit_transform(train_data)
        pcan_test = pcan.transform(test_data)
        GMM = GaussianMixture(n_components=gmm, covariance_type=cov,random_state=rand)
        positive = pcan_train[train_labels==1]
        negative = pcan_train[train_labels==0]       
        GMM_positive = GMM.fit(positive)
        positive_score = GMM_positive.score_samples(pcan_test)
        GMM_negative = GMM.fit(negative)
        negative_score = GMM_negative.score_samples(pcan_test)
        predictions = positive_score > negative_score
        accuracy = float(np.sum(test_labels==predictions))/len(test_labels)
        return accuracy
    
    def computeParams(cov,pca,gmm):
        '''Helper function to compute the total number of parameters'''
        pars = {'spherical':1,'diag':pca,'tied':0.5*pca*(pca+1),'full':0.5*pca*(pca+1)}
        if cov is 'tied':
            return (pca*gmm+pars['tied'])*2
        else:
            # one covariance matrix estimated for all mixture components
            return ((pca+pars['tied'])*gmm)*2
                
    
 # ------------------- Main --------------------------- #

    # define parameter space
    rand_seed = 20
    covariances = ['spherical','diag','tied','full']
    pca_list = np.arange(1,25) 
    gmm_list = np.arange(1,25) 
    results = {} # create empty dictionary for saving the results in
    # loop over the parameter space
    for cov in covariances:
        for pca in pca_list:
            for gmm in gmm_list:
                # Compute number of params 
                parameters = computeParams(cov,pca,gmm)
                if parameters <= 50: # only compute accuracy when # of parameters are < 50
                    # create dictionary with parameter and accuracy as key-value pair 
                    accuracy = fitGMM(cov,pca,gmm,rand_seed)
                    dict_key = str(pca)+","+str(gmm)+","+cov+","+str(parameters)
                    results[dict_key] = accuracy
                
    temp = '|{0:10}|{1:15}|{2:18}|{3:10}|{4:12}|{5:8}|'
    
    # Sort the results dictionary from highest accuracy to smallest
    sorted_results = sorted(results, key=results.get, reverse=True)
    print(temp.format('Model Rank','PCA Components','Mixture Components','Covariance','Total Params','Accuracy'))
    print('-' * 96)
    
    for score_rank,item in enumerate(sorted_results):
        pca,gmm,cov,par = item.split(',')
        print(temp.format(score_rank+1,pca,gmm,cov,int(float(par)),round(results[item],3)))

    
### STUDENT END ###

P6()
```

The models are sorted in decreasing order of accuracy and the hyperparameters that give us the best accuracy are PCA =4, GMM Components = 3, Covariance Type = "Tied" (Total parametes = 44)
